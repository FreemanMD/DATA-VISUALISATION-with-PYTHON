**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
# with seaborn

plt.figure(figsize=(10,6))
plt.suptitle("Box Plot - Age distribution by gender - separeted by manner of death")

sns.boxplot(x= df.gender, y=df.age, data=df, hue= "manner_of_death"	);
```

## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn

plt.figure(figsize=(10,6))
plt.suptitle("Swarm Plot - Age distribution by gender - separeted by manner of death")

sns.swarmplot(x= df.gender, y=df.age, data=df, hue= "manner_of_death"	);
```
